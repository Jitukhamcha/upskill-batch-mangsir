```python
import pandas as pd
data_visu=pd.read_csv('C:/Users/dell/Desktop/adsaada/upskill-batch-mangsir/datas/dummy_dataset_large.csv')
data_visu


import matplotlib.pyplot as plt
# plt.figure(figsize=(9,20))
plt.plot(data_visu['Age'],data_visu['Name'],marker='s',linestyle=':',color="r",label='Age',markersize=8,markerfacecolor='b',linewidth=4)
plt.plot(data_visu["Score"],data_visu['Name'],marker='o',linestyle="-",color="b",label='Score',ms=9,linewidth=5,markerfacecolor='r')
plt.title("Age of people along score")
plt.xlabel('Age')
plt.ylabel('Name')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
Kalimati_data=pd.read_csv('C:/Users/dell/Desktop/adsaada/upskill-batch-mangsir/datas/kalimati_tarkari_dataset_cleaned_2013_to_2021.csv')
Kalimati_data

Kalimati_data.tail(40)
Kalimati_data[Kalimati_data['Commodity'].duplicated()].count
Kalimati_data[Kalimati_data["Commodity"]=='Ginger']

plt.figure(figsize=(200,12))
Ginger=Kalimati_data[(Kalimati_data["Commodity"]=='Ginger') & (Kalimati_data['Date']<'2013-6-10')]


plt.plot(Ginger["Date"],Ginger['Average'],marker='s',linewidth=2,ms=6,markerfacecolor='red',label='Average(Ginger)')
plt.plot(Ginger["Date"],Ginger['Minimum'],marker='o',linewidth=2,ms=6,markerfacecolor='blue',color="red",label='Minimum(Ginger)')


plt.title("Ginger average/minimum along potato red average price",fontsize=20)
plt.xlabel("Date",fontsize=12)
plt.ylabel("Average price")
plt.grid()
plt.legend()
# plt.tight_layout()
plt.show()
```

```python
Kalimati_data[Kalimati_data["Commodity"]=='Potato Red']
Potato_red=Kalimati_data[(Kalimati_data['Commodity']=='Potato Red')&(Kalimati_data['Date']<'2013-11-10')]
plt.figure(figsize=(200,17))
plt.bar(Potato_red["Date"],Potato_red['Maximum'],linewidth=2,edgecolor='red',hatch='o',color='green',label='Average(Potato_red)')
plt.title('Potato -red Maximum price between 2013-06-00 to 2013-11-10')
plt.xlabel("Date")
plt.ylabel("Maximum  price")
plt.show()
```

```python
#histogram
Kalimati_data['Commodity']
Kalimati_data['Commodity'].duplicated()
# Kalimati_data.columns
Kalimati_data[Kalimati_data.SN==197156].count
Garlic_dry_nepali=Kalimati_data[(Kalimati_data['Commodity']=="Garlic Dry Nepali")&(('2013-06-16'<Kalimati_data['Date']) & (Kalimati_data['Date']<'2013-10-21'))]
Garlic_dry_nepali
plt.hist(Garlic_dry_nepali['Maximum'],bins=6,color="red",edgecolor="blue",density=True) #,histtype='step',range=(),
plt.title("number of times Price of Garlic(Nepal) distribution when the price rise")
plt.xlabel("Maximum price of garlic dry Nepal")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter
plt.figure(figsize=(200,12))
plt.scatter(Ginger['Date'],Ginger['Average'],marker='s',s=90,color='blue',alpha=0.9)
plt.title("Ginger Average price as time moves ",fontweight='bold')
plt.xlabel('Date',style="italic")
plt.ylabel('Average price')
plt.grid(True)
plt.show()
```
